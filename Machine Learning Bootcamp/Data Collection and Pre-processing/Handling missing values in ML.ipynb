Imputation :
    1. Mean
    2. Median
    3. Mode

[1,2,3,3,4,5] Mean - sum/n Median - Middle value if odd, avg of middle if even, Mode - most occurence

```python
fig = plt.figure(figsize = (8,8))
sns.distplot(placement_df.salary)
```

We can't use mean since our data is positively skewed toward right.. if normally dist > mean else try ti impute with Median or Mode 

```python
placement_df['salary'].fillna(placement_df['salary'].median(), inplace = True)
```
